```python
# Reading the diamonds dataset from a CSV file into a DataFrame
df = pd.read_csv('/kaggle/input/diamonds/diamonds.csv')
```

```python
# Importing seaborn for data visualization and analysis
import seaborn as sns
# Importing LinearRegression model to create a linear model for predictions
from sklearn.linear_model import LinearRegression
# Importing train_test_split to split the data into training and testing sets
from sklearn.model_selection import train_test_split
```

```python
# Displaying the first 3 rows of the DataFrame to understand the structure and sample data
df.head(3)
```

```python
# Converting categorical columns ("cut", "color", "clarity") into numerical dummy variables
# drop_first=True avoids multicollinearity by removing the first category from each column
df = pd.get_dummies(df, columns=["cut", "color", "clarity"], drop_first=True)
```

```python
# Displaying the first 3 rows of the DataFrame to understand the structure and sample data
df.head(3)
```

```python
# Dropping the 'Unnamed: 0' column, likely an unnecessary index column added during data import
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.head(3)
```

```python
# Defining the target variable 'y' as the 'price' column
y = df[['price']]
# Defining the feature set 'x' by removing the 'price' column
x = df.drop("price", axis=1)
```

```python
# Splitting the data into training and testing sets
# 'train_size=0.76' allocates 76% of the data for training, 24% for testing
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=19, train_size=0.76)
```

```python
# Initializing and training the linear regression model on the training data
lm = LinearRegression()
model = lm.fit(x_train, y_train)
```

```python
# Evaluating the model performance by calculating the R-squared score on the test set
# This indicates how well the model generalizes to unseen data
model.score(x_test, y_test)
```

```python
# Evaluating the model performance on the training set by calculating the R-squared score
# This score indicates how well the model fits the data it was trained on
model.score(x_train, y_train)
```
